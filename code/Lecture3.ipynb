```python
from thinkstats2 import Pmf, Cdf, thinkplot
from dice import Dice

def cdf_example():
    pmf = Pmf(range(1,7))
    three_dice = pmf + pmf + pmf
    cdf = three_dice.MakeCdf()
    thinkplot.Cdf(cdf)

dice = Dice({4: 2, 8:3, 10: 2, 6:6, 12: 4, 20:4})
thinkplot.Hist(dice)
```
