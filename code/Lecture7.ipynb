```python
import soccer
import numpy
import thinkplot


hypos = numpy.linspace(0, 12, 201)
suite = soccer.Soccer(hypos)

thinkplot.Pdf(suite, label='prior')
print('prior mean', suite.Mean())

#construct a prior using a pseudo-observation
suite.Update(69)
thinkplot.PrePlot(4)
thinkplot.Pdf(suite, label='prior 2')
print('after one goal', suite.Mean())

#update with real data
suite.Update(11)
thinkplot.Pdf(suite, label='posterior 1')

suite.Update(12)
thinkplot.Pdf(suite, label='posterior 2')

suite.Update(1)
thinkplot.Pdf(suite, label='posterior 3')

suite.Update(2)
thinkplot.Pdf(suite, label='posterior 4')

suite.Update(3)
thinkplot.Pdf(suite, label='posterior 5')

suite.Update(40)
thinkplot.Pdf(suite, label='posterior 6')

suite.Update(10)
thinkplot.Pdf(suite, label='posterior 6')

thinkplot.Show()
```
