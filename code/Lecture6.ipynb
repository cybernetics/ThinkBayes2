```python
from __future__ import print_function, division

import numpy
import thinkbayes2
import thinkplot

from electorate import Electorate
hypos = numpy.linspace(0, 100, 101)
suite = Electorate(hypos)

thinkplot.Pdf(suite, label='prior')

data = 1.1, 3.7, 53
suite.Update(data)

thinkplot.Pdf(suite, label='posterior')
thinkplot.Show()
print(suite.Std())
print(suite.Mean())

print(suite.ProbLess(50))
```

```python
thinkplot.Pdf(suite, label='prior')
data = -2.3, 4.1, 49
suite.Update(data)

thinkplot.Pdf(suite, label='posterior')
thinkplot.Show()
print(suite.Std())
print(suite.Mean())
print(suite.ProbLess(50))
```
